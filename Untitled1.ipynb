```python
data["Gender"].value_counts()
```

```python
data["Gender"].value_counts().plot(kind="pie",autopct="%0.2f%%")
plt.show()
```

```python
 data.groupby("Gender")["Amount"].sum()
```

```python
data.groupby("Gender")["Amount"].sum().plot(kind="pie",autopct="%0.2f%%")
plt.show()
```

```python
data["Gender"].value_counts().plot(kind="bar",color=("pink","blue"))
plt.title("Coustomer divided by Gender")
plt.show()
```

```python
data["Age Group"].value_counts().plot(kind="bar")
plt.title("Coustomer divided by age")
plt.show()
```

```python
data.groupby("Age Group")["Amount"].sum().plot(kind="bar")
plt.title("Coustmer divided by spending & Age")
plt.show()
```

```python
data["Marital_Status"].value_counts().plot(kind="pie",autopct="%0.0f%%")
plt.title("Marid v Unmarid")
plt.show()
```

```python
data["State"].value_counts().plot(kind="bar")
plt.title("Coustmer from different state")
plt.show()
```

```python
data["Amount"].plot(kind="hist",bins=range(0,25000,2500),edgecolor="black")
plt.xticks(range(0,25000,2500))
plt.show()
```

```python
data["Occupation"].value_counts().plot(kind="bar")
plt.show()
```
